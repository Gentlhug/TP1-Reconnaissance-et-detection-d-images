# TP1 Reconnaissance et détection d'images
## Les bases

### 0. Importer les librairies

```python
import cv2 as cv # Pour installer "opencv", exécuter la commande suivante : pip3 install opencv-python
import matplotlib.pyplot as plt # Pour installer "matplotlib", exécuter la commande suivante : python -m pip install -U matplotlib
```

### 1. Importer l'image de votre choix
*(moi c'est une mante religieuse)*

```python
img = cv.imread('img/mante.jpeg')
```

```python
img = cv.cvtColor(img, cv.COLOR_BGR2RGB) # On conserve les couleurs originales de l'image
```

#### 1.1. Afficher l'image avec matplotlib

```python
plt.imshow(img)
plt.axis('off') # On enlève les axes des absysses et des ordonnées (facultatif)
plt.title('Mante religieuse styley') # On définit un titre (facultatif)
plt.show()
```

#### 1.2. Afficher l'image avec OpenCV

```python
cv.imshow('Mante religieuse styley', img)
k = cv.waitKey(0)
```

### 2. Définir l'image en noir et blanc

```python
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

#### 2.1. Afficher l'image avec OpenCV

```python
cv.imshow('Mante religieuse styley old', img_gray)
k = cv.waitKey(0)
```

#### 2.2. Afficher l'image avec matplotlib

```python
plt.imshow(img_gray, cmap='gray')
plt.axis('off')
plt.title('Mante religieuse styley old')
plt.show()
```

### 3. Convertir l'image en noir et blanc, en format png (par exemple)

```python
cv.imwrite('img/mante_gray.png', img_gray)
```

### 4. Redimensionner l'image

```python
img_resized = cv.resize(img, (300, 100), cv.INTER_LINEAR)

# On affiche l'image avec matplotlib
plt.imshow(img_resized)
plt.axis('off')
plt.title('Mante religieuse styley VladimirPoutineMode')
plt.show()
```

### 5. Rogner l'image

#### 5.0. On crée notre fonction permettant de rogner notre image (facultatif)

```python
def crop_img(img, x, y, w, h):
    cropped_img = img[y:y+h, x:x+w]
    return cropped_img
```

#### 5.1. On affiche les axes de notre image afin de cibler la zone de rognage

```python
plt.imshow(img)
plt.title('Mante religieuse styley')
plt.show()
```

*En l'occurence ici, je veux cibler la tête (x=210 et y=50) d'une largeur et d'une hauteur de 400px*

#### 5.2. On affiche notre image rognée

```python
x = 210
y = 50
width = 400
height = 400

img_cropped = crop_img(img, x, y, width, height)

# On affiche l'image avec matplotlib
plt.imshow(img_cropped)
plt.axis('off')
plt.title('Mante religieuse styley cropped')
plt.show()
```

### 6. Filtre de seuillage

```python
ret, img_thresholded = cv.threshold(img, 200, 255, cv.THRESH_BINARY)
plt.imshow(img_thresholded)
plt.axis('off')
plt.title('Mante religieuse styley dark')
plt.show()
```

### 7. Filtre de flou gaussien

```python
img_blur = cv.GaussianBlur(img, (9, 9), 5)
plt.imshow(img_blur)
plt.axis('off')
plt.title('Mante religieuse styley flou')
plt.show()
```

### 8. Filtre de Canny

```python
img_canny = cv.Canny(img, 30, 40)
plt.imshow(img_canny)
plt.axis('off')
plt.title('Mante religieuse styley canny')
plt.show()
```
